# Task 1: Write a dynamic function to find the derivative of any function f(x)

````python

# Function to calculate the derivative
def dy_dx(function_str):
    """
    Calculate the derivative of a given function string with respect to x using SymPy.

    Parameters:
    - function_str (str): A string representation of the function f(x) whose derivative is to be computed.

    Returns:
    - str: A string representation of the derivative of the input function with respect to x.

    """
    # Define x as a symbolic variable
    x = sp.symbols('x')

    # Convert the input function string to a SymPy expression
    function = sp.sympify(function_str)

    # Compute the derivative of the function with respect to x
    derivative = sp.diff(function, x)

    # Return the derivative as a string representation
    return str(derivative)


# Example usage and test
if __name__ == "__main__":
    # Example function string
    function_str = 'x**2 + 3*x - 5'

    # Calculate the derivative using dy_dx function
    derivative = dy_dx(function_str)

    # Print the original function and its derivative
    print(f"Original function: {function_str}")
    print(f"Derivative: {derivative}")
````

#Task 3: Plot a graph of the Equation must have 4 points (2 minimas and 2 maximas)

````python

# Define the symbolic variable and polynomial function
x = sp.symbols('x')
f = x**5 - 10*x**3 + 15*x

# Compute the first and second derivatives
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)

# Solve for critical points
critical_points = sp.solve(f_prime, x)
critical_points = [point.evalf() for point in critical_points]

# Determine the nature of each critical point
critical_points_nature = []
for point in critical_points:
    concavity = f_double_prime.subs(x, point)
    if concavity > 0:
        nature = 'local minimum'
    elif concavity < 0:
        nature = 'local maximum'
    else:
        nature = 'inflection point'
    critical_points_nature.append((point, nature))

# Include endpoints of the interval for global extrema analysis
interval_endpoints = [-3, 3]
critical_points.extend(interval_endpoints)

# Evaluate function at critical points and endpoints
f_lambdified = sp.lambdify(x, f, modules=['numpy'])
values_at_points = {point: f_lambdified(float(point)) for point in critical_points}

# Determine global maxima and minima
global_max_point = max(values_at_points, key=values_at_points.get)
global_min_point = min(values_at_points, key=values_at_points.get)
global_max_value = values_at_points[global_max_point]
global_min_value = values_at_points[global_min_point]

# Generate x values and corresponding y values for the plot
x_vals = np.linspace(-3, 3, 400)
y_vals = f_lambdified(x_vals)

# Plot the polynomial function
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$f(x) = x^5 - 10x^3 + 15x$')
plt.title('Polynomial Function with Global and Local Extrema')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Highlight the critical points
for point, nature in critical_points_nature:
    if point != global_max_point and point != global_min_point:
        plt.plot(float(point), f_lambdified(float(point)), 'ro' if 'max' in nature else 'bo')
        plt.text(float(point), f_lambdified(float(point)), f'{nature}\n({float(point):.2f}, {f_lambdified(float(point)):.2f})',
                 horizontalalignment='right' if 'max' in nature else 'left',
                 verticalalignment='bottom' if 'max' in nature else 'top')

# Highlight global maxima and minima
plt.plot(float(global_max_point), global_max_value, 'go', label='Global Maxima')
plt.text(float(global_max_point), global_max_value, f'Global Maxima\n({float(global_max_point):.2f}, {global_max_value:.2f})',
         horizontalalignment='left', verticalalignment='bottom')

plt.plot(float(global_min_point), global_min_value, 'mo', label='Global Minima')
plt.text(float(global_min_point), global_min_value, f'Global Minima\n({float(global_min_point):.2f}, {global_min_value:.2f})',
         horizontalalignment='right', verticalalignment='top')

plt.legend()
plt.show()
````

# Plotting a Polynomial Function with Extrema

Concepts Explained:

Polynomial Function Definition:
- The polynomial function chosen for this example is f(x) = x^5 - 10x^3 + 15x.
- This function is defined using SymPy in Python, allowing for symbolic computation of derivatives and critical points.

# Derivatives and Critical Points:
- First Derivative (f'(x)):
   -  The first derivative is calculated to find critical points where the function's slope is zero. These critical points can indicate potential local maxima or minima.

- Second Derivative (f''(x)):
 -  The second derivative helps determine the concavity of the function at each critical point. A positive second derivative indicates a local minimum, while a negative second derivative indicates a local maximum.

# Finding Extrema:
- Local Extrema:
  -  Using the first and second derivatives, local extrema (minima and maxima) are identified. These points are highlighted on the plot to visually show where the function reaches its highest and lowest values within specific intervals.

- Global Extrema:
  - Global maxima and minima are determined by evaluating the function not only at critical points but also at the endpoints of the interval [-3, 3]. This ensures that all possible extreme values within the specified domain are considered.

# Plotting the Function:
- Graphical Representation:
  - The polynomial function is plotted using matplotlib. The x-axis represents the input values (x), while the y-axis represents the output values (f(x)). The plot includes grid lines for clarity and visual reference.

# Annotation and Legend:
- Annotation of Extrema:
  - Each critical point is marked on the plot with a circle ('o'), color-coded to indicate whether it's a local maximum ('red') or local minimum ('blue'). Annotations display the nature of each extremum (local minima or maxima) along with their coordinates.

- Global Extrema Highlighting:
  - Global maxima and minima are additionally highlighted with larger markers ('go' for global maxima and 'mo' for global minima) and labeled accordingly on the plot. This helps distinguish them from local extrema.

# Visualization and Analysis:
-  This graphical representation not only visually illustrates the behavior of the polynomial function but also provides insight into where its highest and lowest points occur.
-  It demonstrates how symbolic computation with SymPy and numerical plotting with matplotlib can be combined effectively to analyze and visualize mathematical functions.


# Task 4: Create separate array of Minimas and maximas

````python

# Define the symbolic variable and polynomial function
x = sp.symbols('x')
f = x**5 - 10*x**3 + 15*x

# Compute the first and second derivatives
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)

# Solve for critical points
critical_points = sp.solve(f_prime, x)

# Convert critical_points to floats for accurate numerical evaluation
critical_points = [float(point.evalf()) for point in critical_points]

# Determine the nature of each critical point
critical_points_nature = []
for point in critical_points:
    concavity = f_double_prime.subs(x, point)
    if concavity > 0:
        nature = 'local minimum'
    elif concavity < 0:
        nature = 'local maximum'
    else:
        nature = 'inflection point'
    critical_points_nature.append((point, nature))

# Filter out local minima and maxima
local_minima = [(point, f.subs(x, point).evalf()) for point, nature in critical_points_nature if nature == 'local minimum']
local_maxima = [(point, f.subs(x, point).evalf()) for point, nature in critical_points_nature if nature == 'local maximum']

# Find global minima and maxima
all_points = local_minima + local_maxima
global_min = min(all_points, key=lambda p: p[1])
global_max = max(all_points, key=lambda p: p[1])

# Include global minima and maxima in their respective arrays
minima = [global_min] + [pt for pt in local_minima if pt != global_min]
maxima = [global_max] + [pt for pt in local_maxima if pt != global_max]

# Printing the results
print(f'Array of Minima: {minima}')
print(f'Array of Maxima: {maxima}')

# Print the critical points and their nature
print("\nCritical Points:")
for point, nature in critical_points_nature:
    print(f"x = {point}")
````
